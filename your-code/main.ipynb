#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

steps_month = fitbit.groupby(by="Months_encoded").median()

x=steps_month.index
y=steps_month["Steps"]
fig, ax1 = plt.subplots(figsize = (12,4))

ax1.plot(x,y, label="Steps/Month (Median)")

ax1.set_title("Fitbit Montly data",fontdict=font)
ax1.legend()
fig.savefig("MEDIAN STEPS BY MONTH_ENCODED.png")

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

steps_WORK_OR_WEEKEND = fitbit.groupby(by="Work_or_Weekend").median()

x=steps_WORK_OR_WEEKEND.index
y=steps_WORK_OR_WEEKEND["Steps"]

fig, ax2 = plt.subplots(figsize = (12,4))

ax2.bar(x,y, label="Steps/Month (Median)")

ax2.set_title("Fitbit Montly data",fontdict=font)
ax2.legend()
fig.savefig("MEDIAN STEPS BY WORK_OR_WEEKEND.png")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
y= fitbit["Steps"]
x=["Minutes Lightly Active","Minutes Very Active","Minutes Sedentary"]
f, ax = plt.subplots(1, 3, figsize=(12, 8))

for i in range(3):
    ax[i].scatter(fitbit[x[i]],y,label= (x[i] + " vs Steps"))
    ax[i].set_title(x[i])
    ax[i].legend()
```
